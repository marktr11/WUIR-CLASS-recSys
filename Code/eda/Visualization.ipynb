```python
# --- Step 2: Calculate the differences (n - o) ---
diff_original = n_attractiveness_ppo_o - o_attractiveness_ppo_o
diff_se_cc = n_attractiveness_ppo_se_cc - o_attractiveness_ppo_se_cc
diff_se = n_attractiveness_ppo_se - o_attractiveness_ppo_se

# --- Step 3: Prepare data for the plot ---
categories = ['Original', 'SE-CC', 'SE']
differences = [diff_original, diff_se_cc, diff_se]

# --- Step 4: Create the bar chart ---
fig, ax = plt.subplots(figsize=(8, 6)) # Create a figure and an axes object

# --- MODIFICATIONS START HERE ---
bar_width = 0.5 # Make bars narrower (default is 0.8)
dark_colors = ['darkblue', 'darkred', 'darkgreen'] # Use darker colors

bars = ax.bar(categories, differences,
              color=dark_colors, # Apply darker colors
              width=bar_width)   # Apply narrower width
# --- MODIFICATIONS END HERE ---

# --- Step 5: Customize the plot with larger fonts ---
# --- MODIFICATIONS START HERE ---
label_fontsize = 15
title_fontsize = 15
tick_fontsize = 15
value_fontsize = 14

ax.set_ylabel('Difference in Attractiveness Value (n - o)', fontsize=label_fontsize) # Set Y-axis label size
ax.set_xlabel('Calculation Context / Category', fontsize=label_fontsize) # Set X-axis label size
ax.set_title('PPO Difference in Attractiveness (n_attractiveness - o_attractiveness)', fontsize=title_fontsize) # Set title size

# Increase tick label size
ax.tick_params(axis='both', which='major', labelsize=tick_fontsize)
# --- MODIFICATIONS END HERE ---

# Add a horizontal line at y=0 for reference
ax.axhline(0, color='grey', linewidth=0.8)

# (Optional) Add data labels on top of the bars with larger font
for bar in bars:
    yval = bar.get_height()
    # Adjust text position based on whether the value is positive or negative
    vertical_alignment = 'bottom' if yval >= 0 else 'top'
    # --- MODIFICATION HERE ---
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.2f}',
             va=vertical_alignment, ha='center',
             fontsize=value_fontsize) # Set data value font size
    # --- MODIFICATION END ---

# --- Step 6: Display the plot ---
plt.tight_layout() # Adjust plot layout to prevent labels overlapping
plt.show()

# Print the calculated differences for verification
print(f"Difference (Original): {diff_original:.2f}")
print(f"Difference (SE-CC): {diff_se_cc:.2f}")
print(f"Difference (SE): {diff_se:.2f}")
```
